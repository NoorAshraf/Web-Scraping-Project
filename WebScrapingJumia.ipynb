```python
from bs4 import BeautifulSoup
import requests
import lxml
from xlwt import *
workbook = Workbook(encoding = 'utf-8')
table = workbook.add_sheet('data')
table.write(0, 0, 'product_page')
table.write(0, 1, 'product_Name')
table.write(0, 2, 'Price range')
table.write(0 ,3, 'Company name')
table.write(0 , 4 , 'Product_Image')
table.write(0 , 5, 'Productrate')
line = 1
for i in range(1 , 4):

  if i == 1:  
   url = 'https://www.jumia.com.eg/mobile-phones/'
   print(url)
  if i != 1:
   url = 'https://www.jumia.com.eg/mobile-phones/'
   url = url + '?page=' + str(i) + '#catalog-listing'
   print(url)
  headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
  }
  html_page = requests.get( url , headers=headers)
  soup = BeautifulSoup(html_page.text , 'lxml')
  links_list = []
  flag = 0
  products = soup.find_all('article',{'class':'prd _fb col c-prd'}) 
  for p in products:
      x = p.find('a')
      urls = 'https://jumia.com.eg' + x['href']
      pro_url = urls
      pro_s_url = requests.get(urls, headers = headers)
      pro_soup = BeautifulSoup(pro_s_url.text , 'lxml')
      product_imgs = p.find('div' , {'class' : 'img-c'})
      pf = product_imgs.find('img').attrs
      pfimg = pf['data-src']
      product_name = pro_soup.find('h1' , {'class':'-fs20 -pts -pbxs'})
      if product_name is not None:
        pr_name = product_name.text
        price_range = pro_soup.find('span' , {'class' : '-b -ltr -tal -fs24'}).text
        company_name_tr = pro_soup.findAll('div' , {'class' : '-pvxs'})
        product_rating = pro_soup.find('a' , {'class':'-plxs _more'}).text
        result = ""
        for cp in company_name_tr:
         n = cp.find('a' , {'class' : '_more'})
         if n is not None:
          result = n.text
          break

        print(pr_name)
        table.write(line,0,urls)
        table.write(line,1,pr_name)
        table.write(line,2,price_range)
        table.write(line,3,result)
        table.write(line,4,pfimg)
        table.write(line,5,product_rating)
        line += 1
 
  print("Done")
  workbook.save('JumiaList.csv')
   
    
     
    
    
 
 
```
